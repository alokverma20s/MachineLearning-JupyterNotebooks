##Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# 2. setup plot
fig , ax = plt.subplots(figsize = (10, 10))

# 3. Plot data
ax.plot(x, y);

# 4. Customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & show (You save the whole figure)
fig.savefig("images/sample-plot.png")
```

### Two options for the subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows =2,
                                             ncols = 2,
                                             figsize = (10, 5))

# plot to each different axis
ax1.plot(x, x/2)
ax2.bar(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax3.scatter(np.random.random(10), np.random.random(10))
ax4.hist(x);
```

```python
# subplot option 2
fig , ax = plt.subplots(nrows= 2,
                        ncols = 2,
                        figsize = (10, 5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (Pyplot vs matplotlib OO method)
* When plotting something quickly, Okay to use the pyplot Method
* When plotting something quickly, Okay to use the OO Method

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50.head()
len(over_50)
```

```python
#pyplot method
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             figsize = (10,6));
```

```python
# OO method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);

# ax.set_xlim([45, 100])
```

```python
# OO Method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50.age,
                       y = over_50.chol,
                       c = over_50.target)
ax.set(title = "Heart Disease ans cholestrol levels",
       xlabel = 'Age',
       ylabel = 'Cholestrol')
#Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal liine
ax.axhline(over_50.chol.mean(),
           linestyle = '--');
```

```python
#Subplot of chol, age thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex = True)
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                   y = over_50['chol'],
                   c = over_50['target'])
#Customize ax0
ax0.set(title = 'Heart Disease and Cholestrol levels',
        ylabel = 'Cholestrol level')
#Add legend to the ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add mean line to the plot
ax0.axhline(over_50['chol'].mean(),
            linestyle = '--')




scatter2 = ax1.scatter(x = over_50.age,
                       y = over_50.thalach,
                       c = over_50.target)

ax1.set(title = 'Heart Disease and Thalach level',
        ylabel = 'Thalach level',
        xlabel = 'Age')

ax1.legend(*scatter2.legend_elements(), title = 'Target')

ax1.axhline(over_50.thalach.mean(), linestyle = '--')


fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```

```python
# Set the plot style
plt.style.use("seaborn-v0_8-whitegrid")

# OO Method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50.age,
                       y = over_50.chol,
                       c = over_50.target,
                       cmap = "winter") # This Changes the color scheme
ax.set(title = "Heart Disease ans cholestrol levels",
       xlabel = 'Age',
       ylabel = 'Cholestrol')
#Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal liine
ax.axhline(over_50.chol.mean(),
           linestyle = '--');
```

```python
#Customizing the y and x axis limitation


#Subplot of chol, age thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex = True)
# Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                   y = over_50['chol'],
                   c = over_50['target'],
                   cmap= 'winter')
#Customize ax0
ax0.set(title = 'Heart Disease and Cholestrol levels',
        ylabel = 'Cholestrol level')
# change the x lim for ax0
ax0.set_xlim([50, 80])
#Add legend to the ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# Add mean line to the plot
ax0.axhline(over_50['chol'].mean(),
            linestyle = '--')




scatter2 = ax1.scatter(x = over_50.age,
                       y = over_50.thalach,
                       c = over_50.target,
                       cmap = 'winter')

ax1.set(title = 'Heart Disease and Thalach level',
        ylabel = 'Thalach level',
        xlabel = 'Age')
ax1.set_ylim([60, 200])

ax1.legend(*scatter2.legend_elements(), title = 'Target')

ax1.axhline(over_50.thalach.mean(), linestyle = '--')


fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
